## Homework 2A 2.10a)

```python
Omega_all = math.factorial(399)/(math.factorial(100)*math.factorial(299))
print("Multiplicity of the whole system: ",Omega_all)
```

```python
q_A = [0]
q_B = [0]

Omega_A = [1]
Omega_B = [1]
Omega_total  = []
for n in range(100):
  n = n+1
  q_A.append(n)
  q_B.append(n)
  Om_A = math.factorial(n+199)/(math.factorial(n)*math.factorial(199))
  Omega_A.append(Om_A)
  Om_B = math.factorial(n+99)/(math.factorial(n)*math.factorial(99))
  Omega_B.append(Om_B)

q_B_rev = q_B[::-1]
Omega_B_rev = Omega_B[::-1]
Omega_total = np.multiply(Omega_A, Omega_B_rev)
```

```python
df = pd.DataFrame({'q_A': q_A,
                   'Omega_A': Omega_A,
                   'q_B': q_B_rev,
                   'Omega_B': Omega_B_rev,
                   'Omega_total': Omega_total})
df
```

```python
fig, ax = plt.subplots()
ax.scatter(df['q_A'], df['Omega_total'], label='qA');
ax.set_xlabel("q_A");
ax.set_ylabel("Omega Total (1x10^95)");
```

```python
fig, ax = plt.subplots()
ax.scatter(df['q_B'], df['Omega_total'], marker='x');
ax.set_xlabel("q_B");
ax.set_ylabel("Omega Total (1x10^95)");
```
